[문제 1]
    
$k=2$일 때, 상태가치함수를 구하시오.

<img src="문제1.png" width="300" height="400"/>

```python
import numpy as np

# 환경 설정
num_rows = 3
num_cols = 3
num_actions = 4 
trap_locations = [(1,3), (3, 1)]
goal_location = (3,3)
discount_factor = 1.0

# 정책 설정 
policy = np.full((num_rows + 2,num_cols + 2,num_actions),0.25)

# 초기화
value_function=np.zeros((num_rows +2 , num_cols +2))

for k in range(2):   # 반복수
    new_value_function=np.zeros((num_rows +2 , num_cols +2))
    for i in range(1,num_rows+1): # 행
        print()
        for j in range(1,num_cols+1): # 열
            print("==================================================================")
            for action in range(num_actions):
                if action==0: next_i,next_j=max(i-1,1),j     #상 
                elif action==2: next_i,next_j=i,max(j-1,0)   # 좌
                elif action==3: next_i,next_j=i,min(j+1,num_cols) # 우
                else : next_i,next_j=min(i+1,num_rows),j     # 하
                
                trap = ((i,j)==(1, 3)) and (action == 0 or action == 3) or ((i,j)==(3, 1)) and (action == 1 or action == 2) 
                
                reward= -0.1 if ((next_i,next_j) in trap_locations or trap) else \
                        (+1 if (next_i,next_j)==goal_location else .0)
                
               
                new_value_function[i][j] += policy[i][j][action]*(reward+(discount_factor*value_function[next_i][next_j])) # 상태가치
                
                print(f"k={k+1}, ({i}, {j}), Action={action}: 식", policy[i][j][action], "*", "(", reward, "+", discount_factor, "*", value_function[next_i][next_j], ")")
    
    value_function=new_value_function.copy()
    
    print(f"\nk={k+1}")
    print("상태가치:")
    print(new_value_function[1:num_rows+1, 1:num_cols+1])
```

```python
import numpy as np

# 환경 설정
num_rows = 3
num_cols = 3
num_actions = 4 
trap_locations = [(1,3), (3, 1)]
goal_location = (3,3)
discount_factor = 1.0

# 정책 설정 
policy = np.full((num_rows + 2,num_cols + 2,num_actions),0.25)

# 초기화
value_function=np.zeros((num_rows +2 , num_cols +2))

for k in range(8):   # 반복수
    new_value_function=np.zeros((num_rows +2 , num_cols +2))
    for i in range(1,num_rows+1): # 행
        print()
        for j in range(1,num_cols+1): # 열
            print("==================================================================")
            for action in range(num_actions):
                if action==0: next_i,next_j=max(i-1,1),j     #상 
                elif action==2: next_i,next_j=i,max(j-1,0)   # 좌
                elif action==3: next_i,next_j=i,min(j+1,num_cols) # 우
                else : next_i,next_j=min(i+1,num_rows),j     # 하
                
                trap = ((i,j)==(1, 3)) and (action == 0 or action == 3) or ((i,j)==(3, 1)) and (action == 1 or action == 2) 
                
                reward= -0.1 if ((next_i,next_j) in trap_locations or trap) else \
                        (+1 if (next_i,next_j)==goal_location else .0)
                
               
                new_value_function[i][j] += policy[i][j][action]*(reward+(discount_factor*value_function[next_i][next_j])) # 상태가치
                
                print(f"k={k+1}, ({i}, {j}), Action={action}: 식", policy[i][j][action], "*", "(", reward, "+", discount_factor, "*", value_function[next_i][next_j], ")")
    
    value_function=new_value_function.copy()
    
    print(f"\nk={k+1}")
    print("상태가치:")
    print(new_value_function[1:num_rows+1, 1:num_cols+1])
```

```python
def max_policy(value_function):
    new_policy=np.zeros((num_rows+2,num_cols+2,num_actions))
    for i in range(1,num_rows+1): # 행
        
        for j in range(1,num_cols+1): # 열
            
            values=[]
            for action in range(num_actions):
                if action==0: next_i,next_j=max(i-1,1),j     #상 
                elif action==2: next_i,next_j=i,max(j-1,0)   # 좌
                elif action==3: next_i,next_j=i,min(j+1,num_cols) # 우
                else : next_i,next_j=min(i+1,num_rows),j     # 하
                
                trap = ((i,j)==(1, 3)) and (action == 0 or action == 3) or ((i,j)==(3, 1)) and (action == 1 or action == 2) 
                
                reward= -0.1 if ((next_i,next_j) in trap_locations or trap) else \
                        (+1 if (next_i,next_j)==goal_location else .0)
                
                values.append(reward+(discount_factor*value_function[next_i][next_j]))
            
            max_value = np.max(values)
            max_value_actions = [action for action, value in enumerate(values) if value == max_value]
            
            prob_for_each_max_action = 1 / len(max_value_actions)
            
            new_policy[i][j] = np.array([prob_for_each_max_action if action in max_value_actions else 0.0 
                                          for action in range(num_actions)])
    
    return new_policy
```
